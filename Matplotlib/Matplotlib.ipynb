```python
# Create a new figure of size 8x6 points, using 100 dots per inch
plt.figure(figsize=(8,6), dpi=80)
plt.subplot(111)
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="--")
plt.plot(X, S, color="red", linewidth=1.0, linestyle="-.")
plt.xlim(-4.0,4.0) 
plt.ylim(-1.0,1.0) 
plt.xticks(np.linspace(-4,4,9,endpoint=True))
plt.yticks(np.linspace(-1,1,5,endpoint=True))
```

```python
# setting limits
plt.plot(X, C, 'r-', X, S, 'g-.')
plt.xlim(X.min()*1.1, X.max()*1.1)
plt.ylim(C.min()*1.1, C.max()*1.1)
```

```python
# setting ticks
plt.plot(X, C, 'r-', X, S, 'g-.')
plt.xlim(-4.0,4.0) 
plt.ylim(-1.0,1.0) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
```

```python
# setting tick labels
plt.plot(X, C, 'r-', X, S, 'g-.')
plt.xlim(-4.0,4.0) 
plt.ylim(-1.0,1.0) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, 1])
```

```python
# moving spines
# Spines are the lines connecting the axis tick marks and noting the boundaries of the data area. They can be placed at arbitrary positions and until now, they were on the border of the axis.
plt.plot(X, C, 'r-', X, S, 'g-.')
plt.xlim(-4.0,4.0) 
plt.ylim(-1.0,1.0) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, +1])
ax = plt.gca()
ax.spines['right'].set_color('none') # discard the top by setting their color to none
ax.spines['top'].set_color('none') # discard the right by setting their color to none
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) # move the bottom and left ones to coordinate 0 in data space coordinates.
```

```python
# adding legend
plt.plot(X, C, 'r-', linewidth=2.5, label="cosine")
plt.plot(X, S, 'g-.',  linewidth=2.5, label="sine")

plt.legend() # 需要在创建图时加上label参数
# 或者plt.legend(['cosine', 'sine'])
```

```python
# OR:
plt.plot(X, C, 'r-', linewidth=2.5)
plt.plot(X, S, 'g-.',  linewidth=2.5)
plt.legend(['cosine', 'sine'], frameon=False) 
# frameon=False是去掉legend的外框
```

```python
# Annotate some Points
plt.plot(X, C, 'r-', X, S, 'g-.')
plt.xlim(-4.0,4.0) 
plt.ylim(-1.0,1.0) 
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$', r'$\pi$'])
plt.yticks([-1, 0, +1])
ax = plt.gca()
ax.spines['right'].set_color('none') # discard the top by setting their color to none
ax.spines['top'].set_color('none') # discard the right by setting their color to none
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0)) # move the bottom and left ones to coordinate 0 in data space coordinates.

t = 2*np.pi/3 # 2.09439
plt.plot([t,t],[0,np.cos(t)], 'b--', linewidth=1.5) # add a straight line
plt.scatter([t,],[np.cos(t),], 50, color='y')
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-60, -40), textcoords='offset points', fontsize=12,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t,t],[0,np.sin(t)], color ='red', linewidth=1.5, linestyle="--")
plt.scatter([t,],[np.sin(t),], 50, color ='red')
plt.annotate(r'$\sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points', fontsize=10,
            arrowprops=dict(arrowstyle='->',
                           connectionstyle='arc3,rad=.2'))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white', edgecolor='None', alpha=0.65))
```

```python
# create some data to use for the plot
dt = 0.001
t = np.arange(0.0, 10.0, dt)
r = np.exp(-t[:1000]/0.05)               # impulse response
x = np.random.randn(len(t))
s = np.convolve(x, r)[:len(x)]*dt  # colored noise

# the main axes is subplot(111) by default
plt.plot(t, s)
plt.axis([0, 1, 1.1*np.amin(s), 2*np.amax(s)])
plt.xlabel('time (s)')
plt.ylabel('current (nA)')
plt.title('Gaussian colored noise')

# this is an inset axes over the main axes
a = plt.axes([.65, .6, .2, .2], facecolor='y')
n, bins, patches = plt.hist(s, 400, normed=1)
plt.title('Probability')
plt.xticks([])
plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='y')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
plt.xticks([])
plt.yticks([])
```

## Bar Plots

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y in zip(X,Y1):
    plt.text(x+0.4, y+0.05, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,Y2):
    plt.text(x+0.4, -y-0.05, '%.2f' % y, ha='center', va= 'top')

plt.ylim(-1.25,+1.25)
plt.xticks([])
plt.yticks([])
```

## Histogram
